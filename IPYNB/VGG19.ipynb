```python
import os
import torch 
import torch.nn as nn
import torch.optim as optim
import torchvision.models as models
from torchvision.transforms.transforms import ToPILImage
from tqdm import tqdm
from torch.utils.data import DataLoader, Dataset, random_split
import matplotlib.pyplot as plt
from torchvision.datasets import ImageFolder
import pandas as pd
from collections import namedtuple
import torchvision.transforms as transforms
import numpy as np
import time
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
import sys
import math
import warnings
warnings.filterwarnings("ignore")
bar_format = '{desc:<5.5}{percentage:3.0f}%|{bar:10}{r_bar}'

path = 'disaster_data'
```

```python
class DisasterData(Dataset):
  def __init__(self, data, augmentation):
    self.data = data
    self.augmentation = augmentation
  
  def __len__(self):
    return len(self.data)
  
  def __getitem__(self, idx):
    image, labels = self.data[idx][0], self.data[idx][1]
    if self.augmentation:
      image = self.augmentation(image)
    return image, labels
```

```python
def train_valid_split(dataset, split_ratio, name1='train', name2='valid'):
    if split_ratio < 0.5:
        valid = int(split_ratio * len(dataset))
        train = int(len(dataset) - valid)
    elif split_ratio > 0.5:
        train = int(split_ratio * len(dataset))
        valid = int(len(dataset) - train)
    else:
        raise Exception("It is not recommened to split any data by 50 %")
    
    train_data, valid_data = random_split(dataset, [train, valid])
    print(f"{name1}: ", len(train_data))
    print(f"{name2}: ", len(valid_data))
    return (train_data, valid_data)
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
num_classes = 6
model = models.vgg19(pretrained=True)
for param in model.parameters():
    param.requires_grad = False

model.classifier[6] = nn.Linear(model.classifier[6].in_features, num_classes)
model = model.to(device)
```

```python
import torch
from sklearn.metrics import confusion_matrix, precision_recall_curve, precision_score, recall_score, f1_score, fbeta_score, r2_score
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import copy

class TorchMetrics:
    def __init__(self, preds, y):
        if type(y) == 'list':
            self.y = y
        else:
            self.y = y.tolist()
        
        if type(preds) == 'list':
            self.preds = preds
        else:
            self.preds = preds.tolist()
        self.num_classes = len(self.y)

    def accuracy(self):
        num_correct =  (np.array(self.preds) == np.array(self.y)).sum()
        return num_correct/len(self.y)

    def getf1_score(self, type, average='micro'):
        if type == 'binary':
            return f1_score(self.y, self.preds)
        else:
            return f1_score(self.y, self.preds, average=average)

    def get_precision_score(self, type, average="micro"):
        if type == 'binary':
            return precision_score(self.y, self.preds)
        else:
            return precision_score(self.y, self.preds, average=average)

    def get_recall_score(self, type,  average="micro"):
        if type == 'binary':
            return recall_score(self.y, self.preds)
        else:
            return recall_score(self.y, self.preds, average=average)

    def f_beta(self, type, beta, average):
        if type == 'binary':
            return fbeta_score(self.y, self.preds, beta)
        else:
            return fbeta_score(self.y, self.preds, beta=beta, average=average)

    def get_confusion_matrix(self, visualise=False, labels=None):
        cm = confusion_matrix(self.y, self.preds)
        if visualise:
            if labels:
                cm_df = pd.DataFrame(cm, index=labels, columns=labels)
            else:
                cm_df = pd.DataFrame(cm)

            plt.figure(figsize=(5,4))
            sns.heatmap(cm_df, annot=True)
            plt.title('Confusion Matrix')
            plt.ylabel('Actal Values')
            plt.xlabel('Predicted Values')
            plt.show()
    
    def get_precision_recall_curve(self):
        precision = dict()
        recall = dict()
        for i in range(self.num_classes):
            precision[i], recall[i], _ = precision_recall_curve(self.y[:,i], self.preds[:, i])
            plt.plot(recall[i], precision[i], lw=2, label='class {}'.format(i))
        plt.xlabel("recall")
        plt.ylabel("precision")
        plt.legend(loc="best")
        plt.title("precision vs. recall curve")
        plt.show()

    def r2_score(self):
        return r2_score(self.y, self.preds)
```

```python
history = {
    'train_loss' : [],
    'valid_loss' : [],
    'train_acc' : [],
    'valid_acc' : [],
    'train_precision' : [],
    'valid_precision' : [],
    'train_recall' : [],
    'valid_recall' : [],
    'train_f1' : [],
    'valid_f1' : []
}
```

```python
def train_model(dataloaders, model, criterion, optimizer, schedular, early_stop_threshold = 5, num_epochs = 10):
  start = time.time()
  best_model_wts = copy.deepcopy(model.state_dict())
    
  best_acc = 0.0
  lowest_loss = float('inf')
  best_precision = 0.0
  best_recall = 0.0
  best_f1 = 0.0
  stop_count = 0

  batch_count = next(iter(dataloaders['train']))[0].shape[0]

  for epoch in range(num_epochs):
    if epoch!= 0 : print()
    print('Epoch {}/{}'.format(epoch, num_epochs - 1))
    print('-' * 10)

    for phase in ['train', 'valid']:
      if phase == 'train':
        model.train()
      else:
        model.eval()

        
      running_loss = 0.0
      running_correct = 0.0
      running_precision = 0.0
      running_recall = 0.0
      running_f1 = 0.0

      batches_till = 0.0
      for inputs, labels in dataloaders[phase]:
        batches_till += 1
        X = inputs.to(device)
        y = labels.to(device)

        optimizer.zero_grad()
        #with torch.set_grad_enabled(phase == 'train'):   # returning bool
        outputs = model(X)
        confidence, predictions = torch.max(outputs, 1)
        loss = criterion(outputs, y)

        met = TorchMetrics(predictions, y)
        sys.stdout.write("\r{}> {} {} {} {} {} {} {} {} {} {} {} {} {}".format(
            "="*int(batches_till/batch_count) , phase, 'after batch:', batches_till, 'loss:', [loss.item()], 'acc:', [met.accuracy()], 'precision:', [met.get_precision_score(type='multiclass')], 'recall: ', [met.get_recall_score(type='multiclass')], 'f1:', [met.getf1_score(type='multiclass')]))
        
        #sys.stdout.write("{} \r{}> {} {} {} {} {} {} {} {}".format(f"after: {batches_till}", "="*int(batches_till//128), 'loss:', loss.item(), 'acc:', met.accuracy(), 'precision:', met.getf1_score(type='multiclass'), 'f1:', met.get_precision_score(type='multiclass')))
        sys.stdout.flush()
        time.sleep(0.5)

        if phase == 'train':
          loss.backward()
          optimizer.step()
          
        running_loss += loss.item() * inputs.size(0)
        running_correct += torch.sum(predictions == y.data)
        metrics = TorchMetrics(predictions, y)
        running_precision += metrics.get_precision_score(type='multiclass')
        running_recall += metrics.get_recall_score(type='multiclass')
        running_f1 += metrics.getf1_score(type='multiclass')

        if phase == 'train':
          history['train_loss'].append(running_loss)
          history['train_acc'].append(running_correct/batches_till)
          history['train_precision'].append(running_precision/batches_till)
          history['train_recall'].append(running_recall/batches_till)
          history['train_f1'].append(running_f1/batches_till)
        
        if phase == 'valid':
          history['valid_loss'].append(running_loss)
          history['valid_acc'].append(running_correct/batches_till)
          history['valid_precision'].append(running_precision/batches_till)
          history['valid_recall'].append(running_recall/batches_till)
          history['valid_f1'].append(running_f1/batches_till)


      if phase == 'train':
        schedular.step()
        
      epoch_loss = running_loss / len(dataloaders[phase].dataset)
      epoch_acc = running_correct / len(dataloaders[phase].dataset)
      epoch_precision = running_precision/batches_till
      epoch_recall = running_recall/batches_till
      epoch_f1 = running_f1/batches_till

      print()
      print('{} Loss: {:.4f} Acc: {:.4f} Precision {:.4f} Recall {:.4f} F1-score {:.4f}'.format(
              phase, epoch_loss, epoch_acc, epoch_precision, epoch_recall, epoch_f1))
      
      # checking test accuracy
      #print(f'TEST ACC: {test_eval(test_path, test_class, transform)}')
        
      if phase == 'valid':
        average_best_metrics = (best_acc + lowest_loss + best_precision + best_recall + best_f1)/5
        average_epoch_metrics = (epoch_acc + epoch_loss + epoch_precision + epoch_recall + epoch_f1)/5
        
        if epoch > 0 and average_epoch_metrics > average_best_metrics:
          print('saving best weights of the model ...')
          best_model_wts = copy.deepcopy(model.state_dict())

        best_acc = max(epoch_acc, best_acc)
        lowest_loss = min(lowest_loss, epoch_loss)
        best_precision = max(best_precision, epoch_precision)
        best_recall = max(best_recall, epoch_recall)
        best_f1 = max(best_f1, epoch_f1)

        if epoch > 0 and epoch_loss > lowest_loss and epoch_acc < best_acc:
          stop_count += 1
        else:
          stop_count = 0
        
      print()
      if stop_count >= early_stop_threshold:
        print(f'Early stopping after epoch {epoch} ...')
        break
    
  time_elapsed = time.time() - start
  print('Training complete in {:.0f}m {:.0f}s'.format(
      time_elapsed // 60, time_elapsed % 60))
  print('Best val Acc: {:4f}'.format(best_acc))
  print('Best val Loss: {:4f}'.format(lowest_loss))
  print('Best val Precision: {:4f}'.format(best_precision))
  print('Best val Precision: {:4f}'.format(best_recall))
  print('Best val Precision: {:4f}'.format(best_f1))

  model.load_state_dict(best_model_wts)
  return model 
```

```python
train_loader = DataLoader(train_dataset, batch_size=128)
valid_loader = DataLoader(valid_dataset, batch_size=128)

dataloaders = {'train':train_loader, 'valid':valid_loader}
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.00175)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=1)
```

```python
model = train_model(dataloaders, model, criterion, optimizer, scheduler, num_epochs=5) 
```

```python
from PIL import Image
def test_eval(test_path, test_class, transform):
  ans = []
  files = test_class.keys()
  labels = test_class.values()
  num_correct = 0
  for file , idx in zip(files, labels):
      path = os.path.join(test_path, file)
      try:
          img = Image.open(path)
          tensor = transform(img)
          temp_new_path = file.split('.')[0] + '.png'
          new_path = os.path.join(test_path, temp_new_path)
          img.save(new_path)
          img_file = Image.open(new_path).convert('RGB')
          tensor_img = transform(img_file)
          tensor = tensor.unsqueeze(0).to(device)
          output = model(tensor)
          confidence, prediction = torch.max(output, 1)
          ans.append(prediction.tolist()[0])
          if prediction == idx:
            num_correct += 1
      except RuntimeError as e:
          print('could not infer dtype of JPEG')
  return ans, (num_correct/len(test_class))
```
